```python
Z = np.random.randn(5,5)
Z[2,3]= np.nan
Z[4,1] = np.nan

Z
```

```python
np.nanvar(Z)
np.nanstd(Z)
np.nanmean(Z)
```

```python
Nnan = np.isnan(Z).sum() # compte le nombre de "nan" dans le tableau
corrnan = Nnan/np.size(Z) # la relation entre "nan" et le restedu tableau
corrnan
```

```python
nn = np.isnan(Z)
Z[nn] = 1
Z
```

```python
np.linalg.det(Z) # Calculer le determinant de la matrice
```

```python
np.linalg.inv(Z) # inverse de la mat
```

```python
np.linalg.pinv(Z)
```

```python
np.linalg.eig(Z) # imprimer les valeurs & vecteurs propres
```

```python
# exo stat

np.random.seed(0)
Y = np.random.randint(0, 100, [10,5])
Y
```

```python
y = Y[:, 0]
num = y - y.mean()
ec_t = y.std()

R = num/ec_t
R
```

```python
for y in Y :
    y = Y[:, 0]
    num = y - y.mean(axis=0)
    ec_t = y.std()
    R = num/ec_t
    print(R)
```
